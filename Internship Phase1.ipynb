```python
#filling the missing values with the mean values
df=df.fillna(df.mean())
```

```python
#function removing the outliers
def rem_out(df,cols,threshold_z):
  for col in cols:
    df=df[np.abs((df[col] - df[col].mean())/df[col].std()) < threshold_z]
  return df
```

```python
outlier_column = ['release_year']
df=rem_out(df,outlier_column,3)
```

```python
df['date_added']=pd.to_datetime(df['date_added'])
df['date_added']
```
